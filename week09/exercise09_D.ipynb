```python
data = [(1, "a@email.com"),
        (2, "b@email.com"),
        (3, "c@email.com")]
schema = ["id", "email"]
df_customers = spark.createDataFrame(data, schema) 

data = [("S1", 1, "C1", "2024-01-01"),
        ("S2", 2, "C1", "2024-01-01"),
        ("S3", 1, "C2", "2024-02-01")]
schema = ["send_id", "customer_id", "campaign_id", "send_date"]
df_campaign_sends = spark.createDataFrame(data, schema) 

data = [("CL1", "S1", "2024-01-02")]
schema = ["click_id", "send_id", "click_date"]
df_clicks = spark.createDataFrame(data, schema) 

from pyspark.sql.functions import lit  # Added import for lit

# q1 Show all campaign sends with customer emails: email, campaign_id, send_date 
df_q1 = df_campaign_sends.join(df_customers, df_campaign_sends.customer_id == df_customers.id, "inner").select(df_customers.email, df_campaign_sends.campaign_id, df_campaign_sends.send_date)
display(df_q1)

# q2 Identify whether each campaign send resulted in a click: send_id, clicked
df_q2 = df_clicks.select(df_clicks.send_id.alias("send_id"), lit(True).alias("clicked"))
df_q2 = df_q2.union(df_campaign_sends.select(df_campaign_sends.send_id.alias("send_id"), lit(False).alias("clicked")))
display(df_q2)

# q3 Show all customers and any campaigns they received: email, campaign_id
df_q3 = df_campaign_sends.join(df_customers, df_campaign_sends.customer_id == df_customers.id, "inner").select(df_customers.email, df_campaign_sends.campaign_id)
display(df_q3)

# q4 find campaign sends that were never clicked: send_id 
df_q4 = df_campaign_sends.join(df_clicks, df_campaign_sends.send_id == df_clicks.send_id, "leftanti").select(df_campaign_sends.send_id)
display(df_q4)

# q5 find customers who have never received a campaign send: email
df_q5 = df_customers.join(df_campaign_sends, df_customers.id == df_campaign_sends.customer_id, "leftanti").select(df_customers.email)
display(df_q5)
 
```
